```python
title_a = pd.read_csv(f'{code[0]}_review_title.csv', encoding='cp949')
title_b = pd.read_csv(f'{code[1]}_review_title.csv', encoding='cp949')
title_c = pd.read_csv(f'{code[2]}_review_title.csv', encoding='cp949')
title_d = pd.read_csv(f'{code[3]}_review_title.csv', encoding='cp949')
title_e = pd.read_csv(f'{code[4]}_review_title.csv', encoding='cp949')
title_f = pd.read_csv(f'{code[5]}_review_title.csv', encoding='cp949')
title_g = pd.read_csv(f'{code[6]}_review_title.csv', encoding='cp949')
title_h = pd.read_csv(f'{code[7]}_review_title.csv', encoding='cp949')
title_i = pd.read_csv(f'{code[8]}_review_title.csv', encoding='cp949')
title_j = pd.read_csv(f'{code[9]}_review_title.csv', encoding='cp949')
```

```python
text_a = pd.read_csv(f'{code[0]}_review_text.csv', encoding='cp949')
text_b = pd.read_csv(f'{code[1]}_review_text.csv', encoding='cp949')
text_c = pd.read_csv(f'{code[2]}_review_text.csv', encoding='cp949')
text_d = pd.read_csv(f'{code[3]}_review_text.csv', encoding='cp949')
text_e = pd.read_csv(f'{code[4]}_review_text.csv', encoding='cp949')
text_f = pd.read_csv(f'{code[5]}_review_text.csv', encoding='cp949')
text_g = pd.read_csv(f'{code[6]}_review_text.csv', encoding='cp949')
text_h = pd.read_csv(f'{code[7]}_review_text.csv', encoding='cp949')
text_i = pd.read_csv(f'{code[8]}_review_text.csv', encoding='cp949')
text_j = pd.read_csv(f'{code[9]}_review_text.csv', encoding='cp949')
```

```python
title_a.drop(columns=['Unnamed: 0'], inplace=True)
title_b.drop(columns=['Unnamed: 0'], inplace=True)
title_c.drop(columns=['Unnamed: 0'], inplace=True)
title_d.drop(columns=['Unnamed: 0'], inplace=True)
title_e.drop(columns=['Unnamed: 0'], inplace=True)
title_f.drop(columns=['Unnamed: 0'], inplace=True)
title_g.drop(columns=['Unnamed: 0'], inplace=True)
title_h.drop(columns=['Unnamed: 0'], inplace=True)
title_i.drop(columns=['Unnamed: 0'], inplace=True)
title_j.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
text_a.drop(columns=['Unnamed: 0'], inplace=True)
text_b.drop(columns=['Unnamed: 0'], inplace=True)
text_c.drop(columns=['Unnamed: 0'], inplace=True)
text_d.drop(columns=['Unnamed: 0'], inplace=True)
text_e.drop(columns=['Unnamed: 0'], inplace=True)
text_f.drop(columns=['Unnamed: 0'], inplace=True)
text_g.drop(columns=['Unnamed: 0'], inplace=True)
text_h.drop(columns=['Unnamed: 0'], inplace=True)
text_i.drop(columns=['Unnamed: 0'], inplace=True)
text_j.drop(columns=['Unnamed: 0'], inplace=True)
```

# 2. Title merge

```python
title_df = pd.merge(title_a, title_b, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_c, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_d, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_e, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_f, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_g, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_h, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_i, left_on='word', right_on='word', how='outer')
title_df = pd.merge(title_df, title_j, left_on='word', right_on='word', how='outer')

title_df.columns = ['word', 'count_1', 'count_2', 'count_3', 'count_4', 'count_5', 'count_6', 'count_7', 'count_8', 'count_9', 'count_10']

title_df.head()
```

```python
title_df.info()
```

```python
# total_count column을 만들기 위해 모든 Null값은 0으로 대체
title_df = title_df.fillna(0)
```

```python
title_df.info()
```

```python
# total_count column 생성
title_df['total_count'] = title_df.sum(axis=1)
```

```python
title_df.head()
```

```python
# 파일화
title_df.to_csv('total_title.csv', index=False, encoding='cp949')
```

# 3. Text merge

```python
text_df = pd.merge(text_a, text_b, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_c, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_d, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_e, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_f, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_g, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_h, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_i, left_on='word', right_on='word', how='outer')
text_df = pd.merge(text_df, text_j, left_on='word', right_on='word', how='outer')

text_df.columns = ['word', 'count_1', 'count_2', 'count_3', 'count_4', 'count_5', 'count_6', 'count_7', 'count_8', 'count_9', 'count_10']

text_df.head()
```

```python
text_df.info()
```

```python
# total_count column을 만들기 위해 모든 Null값은 0으로 대체
text_df = text_df.fillna(0)
```

```python
# total_count column 생성
text_df['total_count'] = text_df.sum(axis=1)
```
